Charger les documents

```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, StorageContext, load_index_from_storage
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings
```

```python
documents = SimpleDirectoryReader("paper").load_data()
```

Initialiser le LLM (TinyLlama) = 2048 tokens, 1,1 milliard de paramètres, Entraîné sur un corpus de textes variés + affiné pour le dialogue

```python
llm = HuggingFaceLLM(
    model_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    tokenizer_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    context_window=2048,
    max_new_tokens=256,
    device_map="auto"
)
```

```python
embed_model = HuggingFaceEmbedding(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
index.storage_context.persist(persist_dir="./storage")
```

```python
query_engine = index.as_query_engine(response_mode="tree_summarize", verbose=True)

questions = [
    "Write a detailed summary of prompting techniques…",
    "What is fine-tuning of language models?",
    "How can LLMs be used for recommendations in e-commerce?",
    "What are multi-modal embeddings and their applications?"
]

for i, question in enumerate(questions, 1):
    print(f"\nQuestion {i}: {question}")
    response = query_engine.query(question)
    print(f"Réponse :\n{str(response)}")
    print("\n" + "="*80)
```
