Creation of Dataframe with the instruction

```python
# preliminary import
import numpy as np
import pandas as pd
from faker import Faker
import calendar
import matplotlib.pyplot as plt
```

```python
fake = Faker()
n_cities = 10

cities = [fake.city() for _ in range(n_cities)]
months = list(calendar.month_abbr)[1:]

n_months = len(months)

temps = np.random.uniform(low=-5, high=35, size=(n_cities, n_months))
df_temps = round(pd.DataFrame(data=temps, index=cities, columns=months), 2)

df_temps
```

```python

annual_avg_temp = df_temps.mean(axis=1)

print("Température moyenne annuelle par ville :")
print(annual_avg_temp)

city_max_temp = annual_avg_temp.idxmax()
max_temp = annual_avg_temp.max()

city_min_temp = annual_avg_temp.idxmin()
min_temp = annual_avg_temp.min()

print(f"\nVille la plus chaude en moyenne annuelle : {city_max_temp} ({max_temp:.2f} °C)")
print(f"Ville la plus froide en moyenne annuelle : {city_min_temp} ({min_temp:.2f} °C)")
```

```python
plt.figure(figsize=(12, 6))

for city in df_temps.index:
    plt.plot(df_temps.columns, df_temps.loc[city], marker='o', label=city)

plt.title("Évolution mensuelle des températures par ville")
plt.xlabel("Mois")
plt.ylabel("Température (°C)")
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```
