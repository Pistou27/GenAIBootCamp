```python
matrix_3x3 = np.random.randint(0, 10, size=(3,3))
print(matrix_3x3)


det = np.linalg.det(matrix_3x3)

print(f"Déterminant de la matrice : {det:.2f}")

if det != 0:
    inverse_matrix = np.linalg.inv(matrix_3x3)
    print("Inverse de la matrice :\n", inverse_matrix)
else:
    print("La matrice n'est pas inversible (déterminant = 0)")
```

Exercice 6

```python
# Productivity scores of employees before the training program
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivity scores of the same employees after the training program
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Your task is to formulate a hypothesis regarding the training program's effectiveness 
# and test it using basic statistical functions in NumPy.
```

```python
#t_statistic, p_value = stats.ttest_1samp(productivity_after, productivity_before.mean())
t_statistic, p_value = stats.ttest_rel(productivity_after, productivity_before)

# Interpreting the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: training program had a significant effect.")
else:
    print("Fail to reject the null hypothesis: no significant effect detected.")
```

Exercise 8 : Time Series Data Manipulation

```python
# Générer une série temporelle journalière pour 2023
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

data = np.random.randn(len(dates))

time_series = pd.DataFrame(data, index=dates)

# Slicing par trimestres
jan_mar = time_series['2023-01-01':'2023-03-31']
apr_jun = time_series['2023-04-01':'2023-06-30']
jul_sep = time_series['2023-07-01':'2023-09-30']
oct_dec = time_series['2023-10-01':'2023-12-31']

# Afficher les résultats
print("\nJanvier à Mars:\n", jan_mar.head())
print("\nAvril à Juin:\n", apr_jun.head())
print("\nJuillet à Septembre:\n", jul_sep.head())
print("\nOctobre à Décembre:\n", oct_dec.head())
```
