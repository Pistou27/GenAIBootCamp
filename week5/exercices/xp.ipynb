```python
def detect_outliers_iqr(df, colonne):
    Q1 = df[colonne].quantile(0.25)
    Q3 = df[colonne].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[colonne] < lower_bound) | (df[colonne] > upper_bound)]
    
    print(f"Colonne : {colonne}")
    print(f"Nombre d'outliers détectés : {outliers.shape[0]}")
    
    return lower_bound, upper_bound

```

```python

lower_bound, upper_bound = detect_outliers_iqr(df, 'ApplicantIncome')
df_clean = df[(df['ApplicantIncome'] >= lower_bound) & (df['ApplicantIncome'] <= upper_bound)]
```

```python
lower_bound, upper_bound = detect_outliers_iqr(df_clean, 'CoapplicantIncome')
df_clean = df_clean[(df_clean['CoapplicantIncome'] >= lower_bound) & (df_clean['CoapplicantIncome'] <= upper_bound)]
```

```python
df_clean = df.drop("Loan_ID", axis=1)
bool_map = {
    "Male": 1, "Female": 0,
    "Yes": 1, "No": 0,
    "Y": 1, "N": 0,

}

df_clean["Gender"] = df_clean["Gender"].map(bool_map)
df_clean["Married"] = df_clean["Married"].map(bool_map)
df_clean["Self_Employed"] = df_clean["Self_Employed"].map(bool_map)
df_clean["Loan_Status"] = df_clean["Loan_Status"].map(bool_map)
df_clean["Education"] = df_clean["Education"].map({"Graduate": 1, "Not Graduate": 0})
df_clean["Property_Area"] = df_clean["Property_Area"].map({"Urban": 1, "Rural": 0})
# La colonne Dependents contient parfois '3+' → il faut la convertir proprement
df_clean["Dependents"] = df_clean["Dependents"].replace("3+", 3).astype(int)

# Pour Property_Area, qui contient plusieurs catégories → on utilise get_dummies
df_clean = pd.get_dummies(df_clean, columns=["Property_Area"], drop_first=True)

print(df_clean.head().T)
print(df_clean.info())
```

```python
corr_matrix = df_clean.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Matrice de Corrélation")
plt.show()
```
